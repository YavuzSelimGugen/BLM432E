```python
import pandas as pd
import numpy as np
```

```python
energy = pd.read_excel('Energy Indicators.xls')
energy = energy[16:243]
```

```python
energy = energy.drop(['Unnamed: 0', 'Unnamed: 1'], axis=1)
energy.reset_index()
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy.set_index("Country")
```

```python
# Nan Temizle
```

```python
energy['Energy Supply'] = energy['Energy Supply'].apply(lambda x: x * 1000000 if type(x) is int else np.NaN)
```

```python
energy['Country'] = energy['Country'].replace(
    {'China, Hong Kong Special Administrative Region':'Hong Kong',
     'United Kingdom of Great Britain and Northern Ireland':'United Kingdom',
     'Republic of Korea':'South Korea','United States of America':'United States',
     'Iran (Islamic Republic of)':'Iran'}
)
energy['Country'] = energy['Country'].str.replace(r" \(.*\)",'')
```

```python
gdp = pd.read_csv('world_bank.csv')
```

```python
# Redefine columns
new_header = gdp.iloc[3]
gdp = gdp[0:]
gdp.columns = new_header
gdp['Country Name'] = gdp['Country Name'].replace(
    {'Korea, Rep.':'South Korea',
     'Iran, Islamic Rep.':'Iran',
     'Hong Kong SAR, China':'Hong Kong'}
)
```

```python
ScimEn = pd.read_excel('scimagojr-3.xlsx')
```

```python
gdp = gdp[['Country Name',2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]]
gdp.columns = ['Country','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']
```

```python
merged1 = energy.merge(gdp,how='inner',left_on='Country',right_on='Country').merge(ScimEn,how='inner',left_on='Country',right_on='Country')
len(merged1.index)
```

```python
merged2 = energy.merge(gdp,how='inner',left_on='Country',right_on='Country').merge(ScimEn[ScimEn['Rank']<16],how='inner',left_on='Country',right_on='Country')
len(merged2.index)
```

```python
merged_all = energy.merge(gdp, how='outer',left_on='Country',right_on='Country').merge(ScimEn,how='outer',left_on='Country',right_on='Country')
```

```python
len(merged_all.index) - len(merged1.index)
```
